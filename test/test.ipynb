```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vid_f = 'output.mp4'
cap = cv2.VideoCapture(vid_f)

V_T = []
i = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    V_T.append(cap.get(cv2.CAP_PROP_POS_MSEC))
    cv2.imshow('frame', frame)
    cv2.imwrite(f'photos/{i}.jpg', frame)
    i += 1
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
data = pd.read_csv('tab.csv')
R_T = np.array(data['field.time_lapse'])/1E9
```

```python
V_T = np.array(V_T)/1E3
```

```python
plt.plot(V_T, [1]*len(V_T), label='video')
plt.plot(R_T, [2]*len(R_T), label='ros')
plt.legend()
```

```python
print('video:', len(V_T), 'ros:', len(R_T))
```
